# STEP 1 — Import Library dan Dataset

```python
import pandas as pd
import numpy as np

url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

df.head()
```

#Q1. Berapa jumlah baris dan kolom pada dataset Titanic? Kolom apa saja yang tersedia?

```python
print("Jumlah baris dan kolom:", df.shape)

print("\nDaftar Kolom:")
print(df.columns.tolist())
```

#Q2. Kolom mana saja yang memiliki missing values? Berapa jumlah dan proporsinya (%)?

```python
missing = df.isnull().sum()

persentase = (missing / len(df) * 100).round(2)

hasil_missing = pd.DataFrame({
    "Jumlah Missing": missing,
    "Persentase (%)": persentase
})

print(hasil_missing[hasil_missing["Jumlah Missing"] > 0])
```

#Q3. Berapa persen penumpang yang selamat (Survived = 1)?

```python
persen_selamat = df["Survived"].mean() * 100

print(f"Persentase penumpang selamat: {persen_selamat:.2f}%")
```

#Q4. Filter penumpang wanita dari kelas 1 (Pclass = 1). Berapa jumlahnya dan berapa rata-rata usianya?

```python
wanita_kelas1 = df[
    (df["Sex"] == "female") &
    (df["Pclass"] == 1)
]

jumlah = wanita_kelas1.shape[0]
rata_usia = wanita_kelas1["Age"].mean()

print("Jumlah wanita kelas 1 :", jumlah)
print("Rata-rata usia :", round(rata_usia,2))
```

#Q5. Gunakan groupby untuk menghitung tingkat keselamatan per kelas penumpang. Kelas mana yang paling tinggi?

```python
keselamatan = df.groupby("Pclass")["Survived"].mean() * 100

print(keselamatan.round(2))
```

#Q6. Analisis tambahan berdasarkan variabel lain yang menarik

```python
survival_gender = df.groupby("Sex")["Survived"].mean() * 100

print(survival_gender.round(2))
```
